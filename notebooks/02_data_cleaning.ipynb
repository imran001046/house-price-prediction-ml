## Structural vs True missing columns concept

- Structural Missing:
   It means that a feature is actually not present for that specific record, for example, a house has no fireplace, usually left blank or NAN while recording the Data so the NAN will be replaced with None in this case

- True Missing:
  It means that a feature was mistakenly or for any other reason was not recorded for that record, like for a house they it must have a front gate, but if the value is missing, it means it was not recorded, so we fill those with median for numeric columns and mode for categorical columns

```python
## filled all structural categorical columns missing values with None 
struct_cat_cols = ['MasVnrType', 'FireplaceQu', 
                   'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
                   'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
df[struct_cat_cols] = df[struct_cat_cols].fillna("None")
```

```python
## Structural Numerical columns that have missing values are filled with zero
struct_num_cols = ['MasVnrArea', 'GarageYrBlt']
df[struct_num_cols] = df[struct_num_cols].fillna(0)
```

```python
## True missing numeric column is filled with median, while categorical with mode, which is the most common value in categorical
df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].median())

df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
```
